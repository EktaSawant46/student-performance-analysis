```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("students_extended.csv")
```

```python
print(df.head())
```

```python
df.isnull().sum()
```

Drop unnamed column

```python
df=df.drop("Unnamed: 0",axis=1)
print(df.head())
```

change WeeklyStudyHours

```python
df["WklyStudyHours"]=df["WklyStudyHours"].str.replace("05-Oct","5-10")
df.head()
```

Gender distribution plot

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

#no. of female students is more than the no. of male students

```python
gb=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
sns.heatmap(gb,annot=True)
plt.figure(figsize=(4,4))
plt.show()
```

#From above chart we can conlude that the education of parents have a good impact on student score


```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
sns.heatmap(gb1,annot=True)
plt.figure(figsize=(4,4))
plt.show()
```

#from above chart we can say marital status of parents have negligible impact on students marks

Distribution of Ethnic Groups

```python
groupA=df.loc[(df["EthnicGroup"]=='group A')].count()
groupB=df.loc[(df["EthnicGroup"]=='group B')].count()
groupC=df.loc[(df["EthnicGroup"]=='group C')].count()
groupD=df.loc[(df["EthnicGroup"]=='group D')].count()
groupE=df.loc[(df["EthnicGroup"]=='group E')].count()

mlist=[groupA["EthnicGroup"], groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(mlist, labels=["groupA","groupB","groupC","groupD","groupE"],autopct="%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```
